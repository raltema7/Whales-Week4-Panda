###  Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios

daily_std = cwa_returns.std()
print(daily_std)
```

###  Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500

# Determine which portfolios are riskier than the S&P 500

riskier = daily_std[daily_std > daily_std["s&p500"]]
print(riskier) # Below Portfolios are Riskier because their std is GREATER than SP500
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)

annualized_std = daily_std * np.sqrt(252)
annualized_std
```

###  Rolling Statistics

### Calculate and plot rolling std for all portfolios with 21-day window

###  Calculate and plot the correlation

```python
# Calculate the correlation

import seaborn as sns
correlation = cwa_returns.corr()
sns.heatmap(correlation, vmin=-1, vmax=1,annot=True,)
correlation
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio

# Calculate variance of S&P 500

# Computing beta

# Plot beta trend

covariance = cwa_returns['TIGER GLOBAL MANAGEMENT LLC'].cov(cwa_returns['s&p500'])
variance = cwa_returns['s&p500'].var()
beta = covariance / variance
beta
```

### Rolling Statistics Challenge: Exponentially Weighted Average

```python
rolling_covariance = cwa_returns['TIGER GLOBAL MANAGEMENT LLC'].rolling(window=60).cov(cwa_returns['s&p500'])
rolling_variance = cwa_returns['s&p500'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 30-Day Beta of TIGER GLOBAL MANAGEMENT LLC')
```

```python
# Use `ewm` to calculate the rolling window

cwa_returns.ewm(halflife=21).std().plot(figsize=(10,10))
```

### Sharpe Ratios 

```python
# Annualized Sharpe ratios 


sharpe_ratios = (cwa_returns.mean() * 252) / (cwa_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot

sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```
